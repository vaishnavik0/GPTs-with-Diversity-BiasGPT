```python
import pandas as pd
import json


file_path = 'asian_bias_dataset.csv'

# Load the dataset
dataset = pd.read_csv(file_path)

print(dataset.head())

# Convert the dataset to the required chat format
chat_format_data = []
for index, row in dataset.iterrows():
    chat_format_data.append({
        "messages": [
            {"role": "user", "content": row['prompt']},
            {"role": "assistant", "content": row['response']}
        ]
    })

output_path = 'asianbias_chat_format.jsonl'
with open(output_path, 'w') as f:
    for entry in chat_format_data:
        f.write(json.dumps(entry) + '\n')

print(f"Data has been converted and saved to {output_path}")
```
